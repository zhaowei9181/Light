我要将  数据少于1000条的（data中）用户剔除掉  怎么写 python代码  我在本地运行

3年是1095  所以  大于1095的

```python
import sys
from pathlib import Path
import pandas as pd

# 导入公共路径
sys.path.append(str(Path("../00_common").resolve()))
from base_path import ROOT

# 读取合并后的数据
data = pd.read_csv(ROOT / "01 数据处理" / "01merged.csv")

# 输出路径
output_path = ROOT / "01 数据处理" / "03filtered.csv"
```

```python
# 统计每个用户的记录数
user_counts = data['uid'].value_counts()

# 条件 1：统计每个用户的总记录数 ≥ 1095
sensing_valid_users = user_counts[user_counts > 1095].index

# 条件 2：每个用户的非空 pam 值数量 ≥ 100
pam_counts = data[~data['pam'].isna()]['uid'].value_counts()
pam_valid_users = pam_counts[pam_counts >= 100].index

# 取交集：满足两个条件的用户
valid_users = sensing_valid_users.intersection(pam_valid_users)

# 保留这些用户的数据
filtered_data = data[data['uid'].isin(valid_users)]

# 可选：保存结果
filtered_data.to_csv(output_path, index=False)

print(f"原始用户数：{data['uid'].nunique()}，保留用户数：{len(valid_users)}")
print(f"处理后数据量：{len(filtered_data)} 行")
```
