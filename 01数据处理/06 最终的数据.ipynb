```python
import sys
from pathlib import Path
import pandas as pd

# 导入公共路径
sys.path.append(str(Path("../00_common").resolve()))
from base_path import ROOT

# 输入和输出路径
input_path = ROOT / "01 数据处理" / "05final.csv"
output_path = ROOT / "01 数据处理" / "06final.csv"
```

```python

# 读取原始数据和性别种族信息
df = pd.read_csv(input_path)



# 计算派生特征
df["motion_time"] = (
    df["act_in_vehicle_ep_0"]
    + df["act_on_bike_ep_0"]
    + df["act_on_foot_ep_0"]
    + df["act_running_ep_0"]
    + df["act_walking_ep_0"]
)

df["still_time"] = df["act_still_ep_0"]
df["device_state"] = df[[f"act_tilting_hr_{i}" for i in range(24)]].sum(axis=1)
df["motion_ratio"] = df["motion_time"] / (df["motion_time"] + df["still_time"])
df["step"] = df["step_ep_0"]
df["unknown_ratio"] = df[[f"act_unknown_hr_{i}" for i in range(24)]].sum(axis=1) / df["motion_time"]

# 重命名一批字段
df = df.rename(columns={
    'audio_amp_mean_ep_0': 'audio_amp_mean', #1
    'audio_amp_std_ep_0': 'audio_amp_std',#2
    'audio_convo_duration_ep_0': 'audio_convo_duration',#3
    'audio_convo_num_ep_0': 'audio_convo_num',#4
    'audio_voice_ep_0': 'audio_voice',#5
    'call_in_duration_ep_0': 'call_in_duration',#6
    'call_in_num_ep_0': 'call_in_num',#7
    'call_miss_num_ep_0': 'call_miss_num',#8
    'call_out_duration_ep_0': 'call_out_duration',#9
    'call_out_num_ep_0': 'call_out_num',#10
    'light_mean_ep_0': 'light_mean',#11
    'light_std_ep_0': 'light_std',#12
    'loc_dist_ep_0': 'loc_dist',#13
    'loc_max_dis_from_campus_ep_0': 'loc_max_dis_from_campus',#14
    'loc_visit_num_ep_0': 'loc_visit_num',#15
    'other_playing_duration_ep_0': 'other_playing_duration',#16
    'other_playing_num_ep_0': 'other_playing_num',#17
    'unlock_num_ep_0': 'unlock_num',#18
    'unlock_duration_ep_0': 'unlock_duration',#19
    'sms_in_num_ep_0': 'sms_in_num',#20
    'sms_out_num_ep_0': 'sms_out_num',#21
})

# 你指定保留的字段列表（共 95 个）
columns_to_keep = ['uid',
'is_ios',
'day',
'gender',
'race',
'is_weekend',
'week',
'motion_time',
'still_time',
'device_state',
'motion_ratio',
'step',
'unknown_ratio',
'audio_amp_mean',
'audio_amp_std',
'audio_convo_duration',
'audio_convo_num',
'audio_voice',
'call_in_duration',
'call_in_num',
'call_miss_num',
'call_out_duration',
'call_out_num',
'light_mean',
'light_std',
'loc_dist',
'loc_max_dis_from_campus',
'loc_visit_num',
'other_playing_duration',
'other_playing_num',
'quality_activity',
'quality_audio',
'quality_light',
'quality_loc',
'unlock_num',
'unlock_duration',
'sms_in_num',
'sms_out_num',
'sleep_duration',
'sleep_end','sleep_start','loc_other_dorm_audio_amp',
'loc_other_dorm_audio_voice','loc_other_dorm_convo_duration','loc_other_dorm_convo_num',
'sleep_heathkit_dur','loc_worship_dur','loc_home_unlock_num','loc_leisure_dur','loc_workout_dur',
'loc_home_still','loc_home_unlock_duration','loc_home_convo_duration','loc_home_convo_num',
'loc_home_dur','loc_home_audio_amp','loc_home_audio_voice','loc_food_unlock_duration',
'loc_food_unlock_num','loc_health_dur','loc_food_dur','loc_food_still','loc_food_audio_voice',
'loc_food_convo_duration','loc_food_convo_num','loc_study_unlock_num','loc_food_audio_amp',
'loc_study_dur','loc_study_still','loc_study_unlock_duration','loc_study_convo_duration',
'loc_study_convo_num','loc_study_audio_amp','loc_study_audio_voice','loc_social_unlock_num',
'loc_social_unlock_duration','loc_social_dur','loc_social_still','loc_social_convo_duration',
'loc_social_convo_num','loc_self_dorm_unlock_num','loc_social_audio_amp','loc_social_audio_voice',
'loc_self_dorm_still','loc_self_dorm_unlock_duration','loc_self_dorm_convo_num','loc_self_dorm_dur',
'loc_self_dorm_audio_amp','loc_self_dorm_audio_voice','loc_self_dorm_convo_duration',
'loc_other_dorm_unlock_duration','loc_other_dorm_unlock_num','loc_other_dorm_dur','loc_other_dorm_still',
'pam']  # 请用你上面列出的完整字段列表替换此行

df_cleaned = df[columns_to_keep]

# 保存新数据
df_cleaned.to_csv(output_path, index=False)
print("✅ 已保存为{output_path}")
```
