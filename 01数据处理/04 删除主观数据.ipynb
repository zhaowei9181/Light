```python
import sys
from pathlib import Path
import pandas as pd

# 导入公共路径
sys.path.append(str(Path("../00_common").resolve()))
from base_path import ROOT

# 读取 CSV 文件
df = pd.read_csv(ROOT / "01 数据处理" / "03filtered.csv")

# 保存结果
output_path = ROOT / "01 数据处理" / "04final.csv"
```

```python
# 需要排除的心理标签关键词（除了 pam）
psych_keywords = [
    'phq4', 'sse3', 'stress', 'social_level', 'avg_ema_spent_time'
]

# 找出所有要删除的列（包含关键词且不是 pam）
cols_to_drop = [col for col in df.columns 
                if any(kw in col for kw in psych_keywords) and col != 'pam']

# 输出被删除的列名
print("以下列将被删除：")
for col in cols_to_drop:
    print(col)

# 删除列
df = df.drop(columns=cols_to_drop)

df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("保存完成：", output_path)
```
