```python
import sys
from pathlib import Path
import pandas as pd

# 导入公共路径
sys.path.append(str(Path("../00_common").resolve()))
from base_path import ROOT

# 读取 CSV 文件
df = pd.read_csv(ROOT / "01 数据处理" / "04final.csv")

# 保存结果
output_path = ROOT / "01 数据处理" / "05final.csv"
```

```python

# 将形如 20170907 的 day 字段转换为 datetime 类型
df['day'] = pd.to_datetime(df['day'].astype(str), format='%Y%m%d')

# 添加是否为周末（Saturday=5, Sunday=6）
df['is_weekend'] = df['day'].dt.weekday >= 5

# 添加星期几（0=周一，6=周日）
df['week'] = df['day'].dt.weekday

# 保存新文件
df.to_csv(output_path, index=False)

print("已添加 'is_weekend' 和 'week' 列")
```
