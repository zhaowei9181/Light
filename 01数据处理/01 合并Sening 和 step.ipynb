将sensing.csv 和 pam.csv,step.csv ，三个表进行左连接，

以 sensing.csv 为主表；

根据 uid 和 day 两列与 steps.csv 和pam.csv对应；

将 steps.csv 和 pam.csv 中的列合并进来（追加在后面）；

如果匹配不上，新增列值为 NaN。

写出python代码

```python
import sys
from pathlib import Path

# 把 00_common 加入 Python 搜索路径
sys.path.append(str(Path("../00_common").resolve()))

from base_path import ROOT

print(ROOT)
print(ROOT.exists())
```

```python
# 输入数据
INPUT_DATA_DIR = ROOT / "00 源数据"
OUTPUT_DATA=ROOT/ "01 数据处理"

sensing_path = INPUT_DATA_DIR / "sensing.csv"
steps_path = INPUT_DATA_DIR / "steps.csv"
pam_path = INPUT_DATA_DIR / "pam.csv"
demographics_path = INPUT_DATA_DIR / "demographics.csv"

print(sensing_path.exists())
print(steps_path.exists())
print(pam_path.exists())
print(demographics_path.exists())

out_path= OUTPUT_DATA/ "01merged.csv"
```

```python
import pandas as pd

sensing = pd.read_csv(sensing_path)
steps = pd.read_csv(steps_path)
pam = pd.read_csv(pam_path)
demographics = pd.read_csv(demographics_path)
```

```python
# 确保 uid 和 day 格式一致（建议统一为字符串）
for df in [sensing, steps, pam]:
    df['uid'] = df['uid'].astype(str)
    df['day'] = df['day'].astype(str)

demographics['uid'] = demographics['uid'].astype(str)

# 先与 steps 左连接
merged = pd.merge(sensing, steps, on=['uid', 'day'], how='left')

# 再与 pam 左连接
merged = pd.merge(merged, pam, on=['uid', 'day'], how='left')

# 再与 demographics 左连接
merged = pd.merge(merged, demographics, on=['uid'], how='left')

# 可选：保存结果
merged.to_csv(out_path, index=False)

# 显示前几行确认
print(merged.head())
```
