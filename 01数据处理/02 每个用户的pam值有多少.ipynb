帮我计算一下  每个用户的  pam  值  有多少个  

计算每个用户（uid），各自有多少个pam值（pam列不为空）

```python
import sys
from pathlib import Path
import pandas as pd

# 导入公共路径
sys.path.append(str(Path("../00_common").resolve()))
from base_path import ROOT

# 读取合并后的数据
pam = pd.read_csv(ROOT / "01 数据处理" / "01merged.csv")

# 输出路径
out_path = ROOT / "01 数据处理" / "02pam_count_per_user.csv"
```

```python
# 确保 uid 是字符串（可选，避免类型不一致）
pam['uid'] = pam['uid'].astype(str)

# 删除 pam 为空的行，只统计有效 pam
pam_clean = pam.dropna(subset=['pam'])

# 统计每个用户的 pam 数量
pam_counts = pam_clean.groupby('uid')['pam'].count().reset_index()
pam_counts.columns = ['uid', 'pam_count']

# 保存为 CSV 文件
pam_counts.to_csv(out_path, index=False)

# 输出确认
print(pam_counts.head())
```
