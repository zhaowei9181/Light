```python
import sys
from pathlib import Path
import pandas as pd
import os

# 导入公共路径
sys.path.append(str(Path("../00_common").resolve()))
from base_path import ROOT

# 读取原始数据
input_path = ROOT / "01 数据处理" / "06final.csv"

# 输出文件夹
output_dir = ROOT / "01 数据处理" / "07split_by_uid"

# 如果输出文件夹不存在，则创建
output_dir.mkdir(parents=True, exist_ok=True)
```

```python
# 读取数据
df = pd.read_csv(input_path)

# 按照 uid 分组并保存
for uid, group in df.groupby("uid"):
    output_path = os.path.join(output_dir, f"{uid}.csv")
    group.to_csv(output_path, index=False)

print(f"已按用户 uid 分割并保存到文件夹：{output_dir}")
```
