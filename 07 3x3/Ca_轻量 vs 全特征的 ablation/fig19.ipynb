```python
import matplotlib.pyplot as plt

# 小时刻度
hours = [f'hr{str(i).zfill(2)}' for i in range(24)]

# 示例数据（用你真实的数据替换）
mse = [6, 7, 12, 17, 22, 22, 17, 10, 5, 6, 6, 4, 5, 6, 4, 5, 4, 4, 5, 6, 7, 5, 5, 5]
extreme_mse = [1, 2, 6, 16, 22, 22, 11, 4, 1, 1, 1, 1, 2, 1, 1, 1, 1, 1, 2, 3, 2, 1, 1, 1]
accuracy = [77, 75, 72, 68, 64, 64, 70, 75, 78, 78, 77, 75, 76, 79, 78, 80, 79, 80, 78, 79, 80, 78, 79, 78]

# 创建图和主轴
fig, ax1 = plt.subplots(figsize=(14, 6))

# 第二个y轴（共享x轴）
ax2 = ax1.twinx()

# 颜色
mse_color = '#8B0000'
extreme_color = '#006400'
acc_color = '#000080'

# 折线图（左轴）
ax1.plot(hours, mse, color=mse_color, marker='o', label='MSE')
ax1.plot(hours, extreme_mse, color=extreme_color, marker='s', label='ExtremeMSE')

# 折线图（右轴）
ax2.plot(hours, accuracy, color=acc_color, marker='^', label='Accuracy')

# y轴标签
ax1.set_ylabel('MSE / ExtremeMSE', fontsize=12)
ax2.set_ylabel('Accuracy (%)', fontsize=12)

# x轴设置
ax1.set_xlabel('Hour', fontsize=12)
ax1.set_xticklabels(hours, rotation=60)

# 标题
plt.title('MSE, ExtremeMSE, and Accuracy per Hour', fontsize=14, weight='bold')

# 图例合并（两个轴的图例放一起）
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper center', ncol=3)

# 网格
ax1.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig("figs/combined_lines_double_axis.png", dpi=300)
plt.show()
```
